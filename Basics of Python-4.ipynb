# Basics of Python - Part 4 - Functions

In this part of the tutorial, we will see how we define functions in Python. This is done by using a def statement. Computers are very useful doing the repetitive tasks in an efficient way. In previous tutorials, we used control flow statements using for loops and while loops and iterate through some given lists (or other iterable objects). But if we would like to write a program to do several different things at once, definitions would be very useful. 

For example, let's say we want to solve several linear programming problems involving two variables. Instead of writing (or copying and pasting the same code over and over), we could define a function in the most general way, and then plug in the only things changing in the problem. At the end of tutorial 5, you'll see an example of that. Right now, let's start with a simple function and see how functions work.

For more information about functions, please look at the documentation at https://docs.python.org/3/tutorial/controlflow.html#defining-functions

On a side note, there are many functions that are already defined and ready to use in Python. Here is the list of all built-in functions in Python: https://docs.python.org/3/library/functions.html

Let's say we want to define a function that multiplies a given number by 2. 

* We first define the function with def, and give it a name. Inside the paranthesis, there is our variable x. 

* Note that in many problems, we could also define a function with more than one inputs.

* We use the word return to get the result.

```python
def f_1(x):
    return 2*x


#let's print some results:
print(f_1(10))
print(f_1(10.554))
print(f_1([1,-2,2.4]))
lst=['a','b']
print(f_1(lst))
```

When defining functions, the word return is an important part of def. If we would like to see the result only, we could also use print() inside the function. However, there is an important difference here. Let's explore more.

If we use return, the type of the result would be the type of the input. That means if input is an integer, the result is an integer, too.If input is a float, as 10.554 above, the result would be a float. 

Since we are multiplying by an integer, we could even have an input as a list. In the examples above, multiplying a list by 2 creates a new list consisting elements of the originial list twice. So, the type of the result would be a list.

Let's see what happens if we use print instead of return.

```python
def f_2(x):
    print(2*x)

print(f_2(10))

```

What's going on here? Note that when we type print(f_2(10)), we get two results, 20 and None. This is because we ask computer to print twice; one in the function and one outside of the function. So, why do we get None?

Note that, in the def statement, we typed that the function prints what it computes (not returning the result), that is why we see the result 20 first. After this, computer looks for f_2(10) but since def is not written with return, the type of the result is NoneType, and computer prints None accordingly.

```python
print(type(f_2(10)))
```

As it is more clear above, the computer works as follow: First it computes f_2(10), then find out its type and then prints the result. That is why we see both 20 and None in the previous result.

## Defining Functions with Lambda

There is one practical way of defining functions without using return in the def statement. These are lambda statements. It is a short way of defining functions.

```python
f_3= lambda x: 2*x

print(f_3(20))

print(type(f_3))

```

## Map

We can use map with lambda to create new lists based on a given list. For example, we want to multiply every element in a given list by 3. So, we use map to apply the function to the given list. 

* Don't forget to use list() to create a new list. Otherwise, the code will only print the information about the map only. 

* On a side note, you could create a new tuple by using tuple() or a new set by using set() instead of list().


```python
lst=[1,2,3,4,'a',['1','a']]

new_lst=list(map(lambda x: x*3,lst))

new_lst
```

As our first example, we will construct a function that plots a given one-variable function over a domain. We use matplotlib package to plot in Python. We will also be using numpy package of Python for some numerical computations. For more information about these packages please take a look at 

https://numpy.org/devdocs/reference/

and

https://matplotlib.org/3.1.1/contents.html

```python
#Note that if you have an error after running these, you may have to install numpy and matplotlib packages in your computer. 
#For this, you could use the following code. Simply uncomment it, run, and comment it again after it runs.
#pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

Now, we are ready to define our function with three inputs: Set X for x-values, set Y for Y values and name is the name of the function to be printed in the title of the plot. 

We will be using plt.plot(X,Y) to have a line plot of these values. If you would like to see a dotplot, please use plt.scatter(X,Y).


```python
def plot_function(X,Y, name):
    plt.plot(X,Y)
    plt.xlabel('x')  # x-axis label
    plt.ylabel('y')  # y-axis label
    plt.title('Plot of {}'.format(name)) #plot title
    return plt.show() 
```

Since we already defined the plot_function, we don't have to repeat the code if we want to plot different functions. This is the convenience that we get by defining the function in a more general way.

```python
#First, define the range of the x-values. This code produces integer values but it would be enough for our examples below.
X=np.arange(-50,50)

#Let's plot different functions.

plot_function(X,X**3,'y=$x^3$')

plot_function(X,X**2,'y=$x^2$')

plot_function(X,np.sin(X),'$sin(x)$')
```

## Example 2:

In this example, we will cover the economic model derived by economists C. Cobb and P. Douglas to explain the American economic growth  between years 1899-1922. 

This model was first introduced in Section 14.1 of Stewart (p.930). Later in Section 14.3 (pp.962-963), how to obtain the model from given assumptions was also explained. 

Recall that, the Cobb-Douglas model is $P=(1.01)L^{0.75}K^{0.25}$.

```python
def CobbDouglas(L,K):
    return (1.01)*L**(0.75)*K**(0.25)
```

In the table provided in the textbook, the values for the first year, 1899, were given as $P=100$, $L=100$ and $K=100$ and for the last year, 1922, are given as $P=240$, $L=161$ and $K=431$. Let's compare model values with these actual values.

```python
values1899=[100,100,100]
values1922=[240,161,431]

print('In 1899, the model gives {} and the actual value was {}'.format(CobbDouglas(100,100),values1899[0]))
print('-------------------------------')
print('In 1922, the model gives {} and the actual value was {}'.format(CobbDouglas(161,431),values1922[0]))
```

The difference in these numbers is probably due to the estimation and the rounding when constructing the model.
